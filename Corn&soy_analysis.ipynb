```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("./soja_milho - precos.csv", skiprows=7)
df
```

```python
df['Preços - Milho'] = df['Preços - Milho'].str.replace('R\$', '').str.replace(',', '.')
df['Preços - Soja'] = df['Preços - Soja'].str.replace('R\$', '').str.replace(',', '.')
df
```

```python
df['Preços - Milho'] = pd.to_numeric(df['Preços - Milho'])
df['Preços - Soja'] = pd.to_numeric(df['Preços - Soja'])
df
```

```python

media_precos_milho = df['Preços - Milho'].mean()
media_precos_soja = df['Preços - Soja'].mean()

print(df)
print("Média dos preços de milho:", media_precos_milho)
print("Média dos preços de soja:", media_precos_soja)
```

```python
desvio_padrao_amostra = df['Preços - Milho'].std(ddof=1)

print(desvio_padrao_amostra)
```

```python
media_precos_feijao = df['Preços - Milho'].mean()

desvio_padrao_amostra = df['Preços - Milho'].std(ddof=1)

coeficiente_variacao = (desvio_padrao_amostra / media_precos_feijao) * 100

print(coeficiente_variacao)
```

```python
media_precos_feijao = df['Preços - Soja'].mean()

desvio_padrao_amostra = df['Preços - Soja'].std(ddof=1)

coeficiente_variacao = (desvio_padrao_amostra / media_precos_feijao) * 100

print(coeficiente_variacao)
```

```python
quartis = df['Preços - Milho'].describe(percentiles=[.25, .5, .75])

print(quartis)
```

```python
quartis = df['Preços - Soja'].describe(percentiles=[.25, .5, .75])

print(quartis)
```

```python
precoMinimo = df['Preços - Milho'].min()
precoMaximo = df['Preços - Milho'].max()

diferenca_amplitude = precoMaximo - precoMinimo

print(precoMinimo)
print(precoMaximo)
print(diferenca_amplitude)
```

```python
precoMinimo = df['Preços - Soja'].min()
precoMaximo = df['Preços - Soja'].max()

diferenca_amplitude = precoMaximo - precoMinimo

print(precoMinimo)
print(precoMaximo)
print(diferenca_amplitude)
```

```python
plt.hist(df['Preços - Milho'], bins=5, edgecolor='black')

plt.xlabel('Preço do Milho')
plt.ylabel('Frequência')
plt.title('Histograma dos Preços do Milho')

for i, freq in enumerate(plt.hist(df['Preços - Milho'], bins=5, edgecolor='black')[0]):
    plt.text(plt.hist(df['Preços - Milho'], bins=5, edgecolor='black')[1][i], freq, str(int(freq)),
             ha='center', va='bottom')

plt.show()
```

```python
plt.hist(df['Preços - Soja'], bins=5, edgecolor='black')

plt.xlabel('Preço da Soja')
plt.ylabel('Frequência')
plt.title('Histograma dos Preços da Soja')

for i, freq in enumerate(plt.hist(df['Preços - Soja'], bins=5, edgecolor='black')[0]):
    plt.text(plt.hist(df['Preços - Soja'], bins=5, edgecolor='black')[1][i], freq, str(int(freq)),
             ha='center', va='bottom')

plt.show()
```

```python
plt.boxplot(df['Preços - Milho'])

plt.xlabel('Preço do Milho')
plt.ylabel('Valores')
plt.title('Boxplot dos Preços do Milho')

plt.show()
```

```python
plt.boxplot(df['Preços - Soja'])

plt.xlabel('Preço da Soja')
plt.ylabel('Valores')
plt.title('Boxplot dos Preços da Soja')

plt.show()
```

```python
plt.plot(df['Preços - Milho'], marker='o', linestyle='-')

for i, preco in enumerate(df['Preços - Milho']):
    plt.text(i, preco, f'{preco:.2f}', ha='right', va='bottom')

plt.xlabel('Índice')
plt.ylabel('Preço do Milho')
plt.title('Variação dos Preços do Milho')

plt.show()
```

```python
plt.plot(df['Preços - Soja'], marker='o', linestyle='-')

for i, preco in enumerate(df['Preços - Soja']):
    plt.text(i, preco, f'{preco:.2f}', ha='right', va='bottom')

plt.xlabel('Índice')
plt.ylabel('Preço da Soja')
plt.title('Variação dos Preços da Soja')

plt.show()
```
